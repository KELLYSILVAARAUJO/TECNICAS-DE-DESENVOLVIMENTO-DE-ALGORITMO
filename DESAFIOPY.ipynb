```python
# =============================
# SIMULA√á√ÉO DO CHAVEAMENTO DA COPA 2026
# =============================

import random

# 1Ô∏è‚É£ LISTA DE SELE√á√ïES PARTICIPANTES (exemplo simplificado)
selecoes = [
    "Brasil", "Argentina", "Fran√ßa", "Alemanha", "Inglaterra", "Espanha",
    "Portugal", "It√°lia", "Holanda", "B√©lgica", "Uruguai", "Cro√°cia",
    "EUA", "M√©xico", "Canad√°", "Jap√£o", "Coreia do Sul", "Austr√°lia",
    "Marrocos", "Senegal", "Nig√©ria", "Camar√µes", "Egito", "Tun√≠sia",
    "Ar√°bia Saudita", "Ir√£", "Catar", "Costa Rica", "Equador", "Chile",
    "Col√¥mbia", "Peru", "Paraguai", "S√©rvia", "Su√≠√ßa", "Pol√¥nia",
    "Ucr√¢nia", "Turquia", "Esc√≥cia", "Pa√≠s de Gales", "Dinamarca",
    "Su√©cia", "Noruega", "Rep√∫blica Tcheca", "Gana", "Arg√©lia",
    "√Åfrica do Sul", "China", "Nova Zel√¢ndia", "Panam√°"
]

# 2Ô∏è‚É£ GERAR 12 GRUPOS DE 4 TIMES (usando dicion√°rio)
random.shuffle(selecoes)
grupos = {}
for i in range(12):
    grupos[f"Grupo {chr(65+i)}"] = selecoes[i*4:(i+1)*4]

# 3Ô∏è‚É£ EXIBIR OS GRUPOS
print("=== FASE DE GRUPOS ===")
for grupo, times in grupos.items():
    print(f"{grupo}: {', '.join(times)}")

# 4Ô∏è‚É£ SIMULAR CLASSIFICA√á√ÉO (dois primeiros + alguns terceiros)
#    - cada grupo gera 4 posi√ß√µes aleat√≥rias
classificados = []
terceiros = []

for grupo, times in grupos.items():
    random.shuffle(times)
    classificados.extend(times[:2])  # 1¬∫ e 2¬∫ lugar
    terceiros.append(times[2])       # poss√≠vel melhor 3¬∫

# 5Ô∏è‚É£ ESCOLHER 8 MELHORES TERCEIROS
melhores_terceiros = random.sample(terceiros, 8)
classificados.extend(melhores_terceiros)

# 6Ô∏è‚É£ FORMAR AS OITAVAS DE FINAL (32 times)
random.shuffle(classificados)
oitavas = [(classificados[i], classificados[i+1]) for i in range(0, 32, 2)]

# 7Ô∏è‚É£ MOSTRAR OITAVAS DE FINAL
print("\n=== OITAVAS DE FINAL ===")
for jogo in oitavas:
    print(f"{jogo[0]} vs {jogo[1]}")

# 8Ô∏è‚É£ FUN√á√ÉO PARA AVAN√áAR DE FASE
def avancar_fase(jogos, nome_fase):
    vencedores = []
    print(f"\n=== {nome_fase.upper()} ===")
    for t1, t2 in jogos:
        vencedor = random.choice([t1, t2])
        vencedores.append(vencedor)
        print(f"{t1} x {t2} ‚Üí {vencedor} venceu")
    return vencedores

# 9Ô∏è‚É£ ELIMINAT√ìRIAS
quartas = [(oitavas[i], oitavas[i+1]) for i in range(0, len(oitavas), 2)]
vencedores_oitavas = avancar_fase(oitavas, "Oitavas de final")

# QUARTAS
jogos_quartas = [(vencedores_oitavas[i], vencedores_oitavas[i+1]) for i in range(0, 16, 2)]
vencedores_quartas = avancar_fase(jogos_quartas, "Quartas de final")

# SEMIFINAIS
jogos_semis = [(vencedores_quartas[i], vencedores_quartas[i+1]) for i in range(0, 8, 2)]
vencedores_semis = avancar_fase(jogos_semis, "Semifinais")

# FINAL
final = [(vencedores_semis[0], vencedores_semis[1])]
campeao = avancar_fase(final, "Final")

print(f"\nüèÜ CAMPE√ÉO DA COPA 2026: {campeao[0]} üèÜ")
```

```python
#exercicio 24/10/2025
#nome: KELLY DA SILVA ARAUJO
#TECNICAS DE DESENVOLVIMENTO DE ALGORITMO

import random

# 1. Gerar 48 sele√ß√µes fict√≠cias
selecoes = [f"Selecao_{i+1}" for i in range(48)]

# 2. Dividir em 12 grupos de 4 sele√ß√µes
grupos = {}
for i in range(12):
    grupos[f"Grupo_{chr(65+i)}"] = selecoes[i*4:(i+1)*4]

# 3. Simular pontua√ß√£o aleat√≥ria para cada sele√ß√£o nos grupos
pontuacoes = {}
for grupo, times in grupos.items():
    pontuacoes[grupo] = {time: random.randint(0, 9) for time in times}

# 4. Classificar os dois primeiros de cada grupo
classificados_diretos = []
terceiros_colocados = []

for grupo, scores in pontuacoes.items():
    ordenados = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    classificados_diretos.extend([ordenados[0][0], ordenados[1][0]])
    terceiros_colocados.append((ordenados[2][0], ordenados[2][1]))

# 5. Selecionar os 8 melhores terceiros colocados
melhores_terceiros = sorted(terceiros_colocados, key=lambda x: x[1], reverse=True)[:8]
classificados_gerais = classificados_diretos + [time for time, _ in melhores_terceiros]

# 6. Embaralhar os 32 classificados para formar confrontos
random.shuffle(classificados_gerais)
confrontos_oitavas = [(classificados_gerais[i], classificados_gerais[i+1]) for i in range(0, 32, 2)]

# 7. Exibir confrontos das oitavas de final
print("üîó Confrontos das Oitavas de Final:")
for i, confronto in enumerate(confrontos_oitavas, 1):
    print(f"Jogo {i}: {confronto[0]} vs {confronto[1]}")
```
